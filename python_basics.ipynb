```python
#Write function is_prime(n) that returns True if n is prime. (Hint: check divisors up to int(n**0.5))
def is_prime(n):
    if n <= 1:
        return False
    
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print(is_prime(10)) 
```

```python
#Write def avg(numbers): that returns average of a list.'
def avg(numbers):
    return sum(numbers) / len(numbers)

print(avg([10, 20, 30]))
```
